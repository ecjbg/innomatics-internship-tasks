```python
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge amount must be at least ₹50"
    
    if plan not in valid_plans:
        return "Invalid plan selected"
    
    return "Recharge Successful"

# Retry system
while True:
    amount = int(input("Enter recharge amount: "))
    plan = int(input("Enter plan: "))
    
    result = validate_recharge(amount, plan)
    print(result)
    
    if result == "Recharge Successful":
        break
```

```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")

products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(products)
```

```python
#3
def process_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 45, 70, 55]
result = process_result(marks)

print("Average:", sum(marks)/len(marks))
print("Result:", result)
```

```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)
    
    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)
    
    return fare

while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ")
    
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)
    
    retry = input("Calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```
